```python
import tkinter as tk
from tkinter import messagebox
from reportlab.lib.pagesizes import letter
from reportlab.pdfgen import canvas

# Create tkinter window
window = tk.Tk()
window.title("Candidate Data Entry")

# Create labels and entry fields for each input
tk.Label(window, text="Name:").grid(row=0, column=0)
name_entry = tk.Entry(window)
name_entry.grid(row=0, column=1)

tk.Label(window, text="AICTE ID:").grid(row=1, column=0)
aict_id_entry = tk.Entry(window)
aict_id_entry.grid(row=1, column=1)

tk.Label(window, text="Email:").grid(row=2, column=0)
email_entry = tk.Entry(window)
email_entry.grid(row=2, column=1)

tk.Label(window, text="Phone no:").grid(row=3, column=0)
phone_no_entry = tk.Entry(window)
phone_no_entry.grid(row=3, column=1)

tk.Label(window, text="College Name:").grid(row=4, column=0)
college_name_entry = tk.Entry(window)
college_name_entry.grid(row=4, column=1)



# Function to handle form submission
def submit_form():
    # Validate data
    if not name_entry.get() or not aict_id_entry.get():
        messagebox.showerror("Error", "Name and AICTE ID are required!")
        return
    # If data is valid, generate PDF
    generate_pdf()
    messagebox.showinfo("Success", "PDF generated successfully!")

# Create submit button
submit_button = tk.Button(window, text="Submit", command=submit_form)
submit_button.grid(row=5, column=0, columnspan=2)

window.mainloop()

def generate_pdf():
    name = name_entry.get()
    aict_id = aict_id_entry.get()
    email = email_entry.get()
    phone_no = phone_no_entry.get()
    college_name = college_name_entry.get()
    
   

    pdf_file = f"{name}_application.pdf"
    c = canvas.Canvas(pdf_file, pagesize=letter)
    c.drawString(100, 750, "Candidate Information:")
    c.drawString(100, 730, f"Name: {name}")
    c.drawString(100, 710, f"AICTE ID: {aict_id}")
    c.drawString(100, 710, f"Email: {email}")
    c.drawString(100, 710, f"Phone no.: {phone_no}")
    c.drawString(100, 710, f"College name: {college_name}")
    
 

    c.save()

```
